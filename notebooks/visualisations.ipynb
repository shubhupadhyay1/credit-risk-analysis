```python
# Credit Risk Analysis - Visualization Notebook

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df = pd.read_csv('../data/train.csv')
test_df = pd.read_csv('../data/test.csv')


# 1. Data Distribution - Numerical Features

num_feats = [col for col in train_df.columns if train_df[col].dtype != 'object' and col != 'credit_card_default']

for feature in num_feats:
    plt.figure(figsize=(12, 6))
    sns.histplot(train_df[feature], kde=True, bins=30)
    plt.title(f'Distribution of {feature}')
    plt.show()
```

```python
# 2. Categorical Feature Analysis

cat_feats = [col for col in train_df.columns if train_df[col].dtype == 'object']

for feature in cat_feats:
    plt.figure(figsize=(12, 6))
    sns.countplot(x=train_df[feature])
    plt.title(f'Count Plot of {feature}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# 3. Correlation Heatmap

plt.figure(figsize=(14, 10))
correlation_matrix = train_df[num_feats].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
# 4. Class Imbalance

plt.figure(figsize=(8, 6))
sns.countplot(x=train_df['credit_card_default'])
plt.title('Class Distribution of Target Variable')
plt.show()
```

```python
# 5. Missing Data Analysis

plt.figure(figsize=(12, 8))
sns.heatmap(train_df.isnull(), cbar=False, cmap='viridis')
plt.title('Heatmap of Missing Values')
plt.show()

```
